```python
import yfinance as yf
import pandas as pd

# Gold Futures symbol
symbol = "GC=F"

# Download data
data = yf.download(
    symbol,
    period="7d",
    interval="5m",
    auto_adjust=False
)

# Reset index
data.reset_index(inplace=True)

# Flatten columns if multi-index
if isinstance(data.columns, pd.MultiIndex):
    data.columns = data.columns.get_level_values(0)

# Create required dataframe
df = pd.DataFrame()

# Time column
if "Datetime" in data.columns:
    df["time"] = data["Datetime"]
else:
    df["time"] = data["Date"]

# OHLCV columns
df["open"] = data["Open"]
df["high"] = data["High"]
df["low"] = data["Low"]
df["close"] = data["Close"]
df["volume"] = data["Volume"]

# Save CSV
csv_name = "xauusd_data.csv"
df.to_csv(csv_name, index=False)

print("CSV saved successfully!")
print(df.head())
```
